Q2.19quiz

```python
import numpy as np
mean = 0
sd = 1.4

s = np.random.normal(loc=mean, scale=sd, size=100000)
x = abs(sd - np.mean(s))
y = abs(sd - np.std(s, ddof=1))

print(f"standard deviation = {x}")
print(f"mean = {y}")
```
